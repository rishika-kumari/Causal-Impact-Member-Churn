```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load processed data
df = pd.read_csv("../../data/processed/merged_panel_clean_data.csv")
df["zeit"] = df["jahr"].astype(str) + "Q" + df["quartal"].astype(str)

# ▶️ Plot contribution rate over time
def plot_contribution_trends(sample_size=6):
    funds = df["kasse_clean"].unique()[:sample_size]
    for fund in funds:
        sub = df[df["kasse_clean"] == fund]
        plt.plot(sub["zeit"], sub["zusatzbeitrag"], label=fund)
    plt.xticks(rotation=90)
    plt.title("Contribution Rate Over Time")
    plt.ylabel("Zusatzbeitrag")
    plt.legend()
    plt.tight_layout()
    plt.show()

# ▶️ Contribution increase summary table
def summarize_contribution_increases():
    increases = df[df["treatment_flag"] == True].groupby("kasse_clean").agg(
        count=("treatment_flag", "sum"),
        avg_increase=("zusatzbeitrag", "mean"),
        max_increase=("zusatzbeitrag", "max")
    ).sort_values(by="count", ascending=False)
    print(increases.head(10))

# Run both
if __name__ == "__main__":
    plot_contribution_trends()
    summarize_contribution_increases()
```
